```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

datasets = ["DTD","EuroSAT","GTSRB","MNIST","SVHN"]
```

```python
for ds in datasets:
    pred = pd.read_csv(f"prediction_{ds}.csv",index_col="Unnamed: 0")
    unc_pred = pd.read_csv(f"uncertainty_{ds}.csv",index_col="Unnamed: 0")
    display(unc_pred.head(1))
    unc_pred
    sns.kdeplot(np.log(unc_pred["ent_18"]), label='Entropy 18')
    sns.kdeplot(np.log(unc_pred["ent_101"]), label='Entropy 101')
    sns.kdeplot(np.log(unc_pred["kl_18_101"]), label='KL(res18||res101)')
    sns.kdeplot(np.log(unc_pred["kl_101_18"]), label='KL(res101||res18)')
    sns.kdeplot(np.log(unc_pred["ce_18_101"]), label='CE(res18||res10)')
    sns.kdeplot(np.log(unc_pred["ce_101_18"]), label='CE(res101||res18)')
    plt.title(f'Uncertainty Distribution for {ds}')
    plt.xlabel("")
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
